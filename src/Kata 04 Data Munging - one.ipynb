```python
import pandas as pd
import numpy as np
```

```python
starFunc = lambda x :x.replace('*','')
cvtrs = {1: starFunc,
         2: starFunc,
         12:starFunc}

df = pd.read_fwf('../data/weather.dat', delim_whitespace=True, index_col=0, converters=cvtrs)
df.drop(np.nan, inplace=True)
df.head()
```

```python
df.drop("mo", inplace=True)
df.tail()
```

```python
import itertools
# Why aren't columns splitting cleanly?

def splitSeries(combinedSeries, delimiter=' ', replaceWhat='*', replaceWith=''):
    """
    Grabs first and last. zip(*[]) unpacks the list comprehension
    """
    zz = zip(*[[x[0], x[-1]] for x in combinedSeries.str.strip().str.replace(replaceWhat,replaceWith).str.split(delimiter)])
    t1, t2 = itertools.tee(zz)
    s1 = pd.Series(list(t1)[0], index=df.index)
    s2 = pd.Series(list(t2)[1], index=df.index)
    
    return s1,s2
```

```python
df['mxs'], df['skyc'] = splitSeries(df['MxS SkyC'])
del df['MxS SkyC']
```

```python
df['mnr'], df['avslp'] = splitSeries(df['MnR AvSLP'])
del df['MnR AvSLP']
```

```python
# Clean up headings
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```
